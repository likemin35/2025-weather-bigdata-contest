관측지점정보

방재기상관측(AWS) 지점에 대한 지점번호, 관측 시작일과 종료일, 위·경도 등 이력 정보와 주변 환경 등에 대한 정보

```python
import urllib.parse

base_url = "https://raw.githubusercontent.com/s0nghyunje0ng/2025-weather-bigdata-contest/safety/topic2_safety/data/external/"
encoded_filename = urllib.parse.quote("META_관측지점정보_20250610001229.csv")
url = base_url + encoded_filename
```

```python
import pandas as pd

raw_meta = pd.read_csv(url, encoding="cp949")
raw_meta.head()
```

```python
raw_meta.info()
```

2020년 5월 1일 ~ 2024년 10월 31일 사이에 관측이 이루어진 지점만 필터링

```python
raw_meta["시작일"] = pd.to_datetime(raw_meta["시작일"])
raw_meta["종료일"] = pd.to_datetime(raw_meta["종료일"])

start_cutoff = pd.Timestamp("2020-05-01")
end_cutoff = pd.Timestamp("2024-10-31")

active_meta = raw_meta[(raw_meta["시작일"] <= end_cutoff) & ((raw_meta["종료일"].isna()) | (raw_meta["종료일"] >= start_cutoff))]
active_meta
```

```python
# 지점별로 위도와 경도에 대한 평균과 표준편차 계산
coord_stats = active_meta.groupby("지점")[["위도", "경도"]].agg(["mean", "std"]).reset_index()
coord_stats
```

관측소 위도/경도 표준편차가 거의 없음 → 같은 자리에서 시작/종료를 반복했을 것으로 추정

```python
# 지점별 위도/경도 평균 테이블
coord_mean = active_meta.groupby("지점")[["위도", "경도"]].mean().reset_index()

# 지점, 지점명, 지점주소 정보 중복 제거 후 하나로 정리
station_info = active_meta.drop_duplicates("지점")[["지점", "지점명", "지점주소"]]

# 병합하여 위도/경도 평균 추가
stations_avg_coord = pd.merge(station_info, coord_mean, on="지점", how="left")
stations_avg_coord
```

```python
stations_avg_coord["지점"].unique()
```

우리가 가진 `stn`과 겹치지 않는 지점 3개 발견 ❗ `160`, `296`, `969`

```python
active_meta[active_meta["지점"].isin([160, 296, 969])]
```

- `160`: `159`와 겹침 → 대체

- `296`: `904`와 겹침 → 제거

- `969`: 더이상 사용하지 않는 지점 → 제거

```python
stations_avg_coord["지점"] = stations_avg_coord["지점"].replace(160, 159)
stations_final = stations_avg_coord[~stations_avg_coord["지점"].isin([296, 969])].reset_index(drop=True)
stations_final
```

인접 관측소 매핑

```python
from geopy.distance import geodesic

# 인접 지점 매핑 (각 지점별 가장 가까운 3개 지점 찾기)
neighbor_rows = []

for idx, row in stations_final.iterrows():
    origin_id = row["지점"]
    origin_name = row["지점명"]
    origin_coord = (row["위도"], row["경도"])

    distances = []

    for idx2, row2 in stations_final.iterrows():
        if row["지점"] == row2["지점"]:
            continue
        target_id = row2["지점"]
        target_name = row2["지점명"]
        target_coord = (row2["위도"], row2["경도"])
        distance_km = geodesic(origin_coord, target_coord).km
        distances.append((target_id, target_name, distance_km))

    # 거리순 정렬 후 상위 3개만 저장
    nearest = sorted(distances, key=lambda x: x[2])[:3]
    for neighbor_id, neighbor_name, dist in nearest:
        neighbor_rows.append({
            "지점": origin_id,
            "지점명": origin_name,
            "인접지점": neighbor_id,
            "인접지점명": neighbor_name,
            "거리(km)": round(dist, 4)
        })

neighbors_df = pd.DataFrame(neighbor_rows)
neighbors_df
```

```python
neighbors_dict = neighbors_df.groupby("지점")["인접지점"].apply(list).to_dict()
neighbors_dict
```

```python
import json

# 저장
with open("busan_stn_neighbors.json", "w", encoding="utf-8") as f:
    json.dump(neighbors_dict, f, ensure_ascii=False)
```
