# 날짜 기반 파생 변수

| 변수명 | 설명 |
|---|---|
| year | 연도 |
| month | 월 |
| dayofweek | 요일 |
| dayofweek_sin | 요일 sin |
| dayofweek_cos | 요일 cos |
| is_weekend | 주말 여부 |
| is_holiday | 공휴일 여부 |
| call_count_prev_year | 전년도 인접일(±2일) 평균 신고 건수 |
| has_prev_year_data | 전년도 데이터 존재 여부 (↑ 플래그) |

```python
shape_before_date_features = df_call.shape
```

```python
df_call["tm"] = pd.to_datetime(df_call["tm"])
df_call["year"] = df_call["tm"].dt.year
df_call["month"] = df_call["tm"].dt.month
df_call["dayofweek"] = df_call["tm"].dt.dayofweek
df_call["dayofweek_sin"] = np.sin(2 * np.pi * df_call["dayofweek"] / 7)
df_call["dayofweek_cos"] = np.cos(2 * np.pi * df_call["dayofweek"] / 7)
df_call["is_weekend"] = (df_call["dayofweek"] >= 5).astype(int)
```

```python
import holidays

kr_holidays = holidays.Korea()
df_call["is_holiday"] = df_call["tm"].dt.date.apply(lambda x: x in kr_holidays).astype(int)
```

```python
from datetime import timedelta

df_prev = df_call[["tm", "address_gu", "sub_address", "call_count"]].copy()
df_prev["tm"] += pd.DateOffset(years=1)
df_all = pd.concat([df_prev.assign(tm=df_prev["tm"] + timedelta(days=d)) for d in [-2, -1, 0, 1, 2]])
df_avg = (
    df_all.groupby(["tm", "address_gu", "sub_address"], observed=False)["call_count"]
    .mean()
    .reset_index()
    .rename(columns={"call_count": "call_count_prev_year"})
)
df_call = df_call.merge(df_avg, on=["tm", "address_gu", "sub_address"], how="left")
df_call.loc[df_call["year"] != 2020, "call_count_prev_year"] = (
    df_call.loc[df_call["year"] != 2020, "call_count_prev_year"].fillna(0)  # 신고가 없었던 날
)
df_call["has_prev_year_data"] = df_call["call_count_prev_year"].notna().astype(int)  # 2020년 이전 데이터는 없기 때문에 플래그 사용
```

```python
date_feature_cols = [
    "year", "month", "dayofweek", "dayofweek_sin", "dayofweek_cos", "is_weekend",
    "is_holiday", "call_count_prev_year", "has_prev_year_data"
]

print(f"✅ Date Features Added: {df_call.shape} (+{df_call.shape[1] - shape_before_date_features[1]} cols)")
df_call[date_feature_cols].sample(3)
```

# 날씨 기반 파생 변수

| 변수명 | 설명 |
|---|---|
| is_rainy | 비 여부 |
| rn_day_bin | 강수량 구간화 |
| rn_day_3d_sum | 최근 3일 누적 강수량 |
| apparent_temp | 체감온도 |
| discomfort_index | 불쾌지수 |

```python
shape_before_weather_features = df_call.shape
```

```python
df_call["is_rainy"] = (df_call["rn_day"] > 0).astype(int)
df_call["rn_day_bin"] = pd.cut(
    df_call["rn_day"],
    bins=[-1, 0, 1, 10, 30, 70, np.inf],
    labels=["none", "trace", "light", "moderate", "heavy", "very_heavy"]
)
df_call["rn_day_3d_sum"] = (
    df_call.groupby(["sub_address"], observed=False)["rn_day"]
    .rolling(window=3, min_periods=1)
    .sum()
    .reset_index(level=0, drop=True)
)
```

```python
Ta = df_call[["ta_max", "ta_min"]].mean(axis=1)  # 평균 기온
RH = df_call[["hm_max", "hm_min"]].mean(axis=1)  # 평균 상대 습도

# 체감온도
Tw = (
    Ta * np.arctan(0.151977 * np.sqrt(RH + 8.313659)) +
    np.arctan(Ta + RH) -
    np.arctan(RH - 1.67633) +
    0.00391838 * (RH ** 1.5) * np.arctan(0.023101 * RH) -
    4.686035
)
df_call["apparent_temp"] = (-0.2442 + 0.55399 * Tw + 0.45535 * Ta - 0.0022 * (Tw ** 2) + 0.00278 * Tw * Ta + 3.0)

# 불쾌지수
df_call["discomfort_index"] = ((9 / 5) * Ta - 0.55 * (1 - RH / 100) * ((9 / 5) * Ta - 26) + 32)
```

```python
weather_feature_cols = ["is_rainy", "rn_day_bin", "rn_day_3d_sum", "apparent_temp", "discomfort_index"]

print(f"✅ Weather Features Added: {df_call.shape} (+{df_call.shape[1] - shape_before_weather_features[1]} cols)")
df_call[weather_feature_cols].sample(3)
```

# 기상특보 기반 파생 변수

| 변수명 | 설명 |
|---|---|
| has_alert | 특보 발효 여부 |
| alert_count | 발효된 특보 개수 |
| alert_type | 발효된 특보종류 집합 |
| alert_streak | 연속 특보 발효 일수 |
| alert_type_score | 특보종류별 가중치 합계 |
| alert_level_score | 특보수준별 가중치 합계 |

```python
shape_before_alert_features = df_call.shape
```

```python
df_wrn["발효시각"] = pd.to_datetime(df_wrn["발효시각"])
df_wrn["발효일"] = df_wrn["발효시각"].dt.normalize()

df_wrn_filtered = df_wrn[df_wrn["특보명령"] != "해제"].copy()

level_score_map = {"주의보": 1, "경보": 2}
type_score_map = {
    "강풍": 1, "건조": 1, "태풍": 3, "폭염": 2,
    "폭풍해일": 3, "풍랑": 2, "호우": 2  # 임의 설정
}

df_wrn_filtered["type_score"] = df_wrn_filtered["특보종류"].map(type_score_map)
df_wrn_filtered["level_score"] = df_wrn_filtered["특보수준"].map(level_score_map)

df_agg = (
        df_wrn_filtered.groupby("발효일", observed=False)
        .agg(
            has_alert = ("특보종류", lambda x: 1),
            alert_count = ("특보종류", "count"),
            alert_type = ("특보종류", lambda x: (set(x))),
            alert_type_score = ("type_score", "sum"),
            alert_level_score = ("level_score", "sum")
        )
        .reset_index()
        .sort_values("발효일")
        .reset_index(drop=True)
    )

streaks = []
count = 0
for i in range(len(df_agg)):
    if i == 0:
        count = 1
    else:
        prev = df_agg.loc[i - 1, "발효일"]
        curr = df_agg.loc[i, "발효일"]
        count = count + 1 if (curr - prev).days == 1 else 1
    streaks.append(count)
df_agg["alert_streak"] = streaks
```

```python
df_call = df_call.merge(df_agg, left_on="tm", right_on="발효일", how="left").drop(columns=["발효일"])

df_call["has_alert"] = df_call["has_alert"].fillna(0).astype(int)
df_call["alert_count"] = df_call["alert_count"].fillna(0).astype(int)
df_call["alert_type"] = df_call["alert_type"].apply(lambda x: x if isinstance(x, set) else set())
df_call["alert_streak"] = df_call["alert_streak"].fillna(0).astype(int)
df_call["alert_type_score"] = df_call["alert_type_score"].fillna(0)
df_call["alert_level_score"] = df_call["alert_level_score"].fillna(0)
```

```python
alert_feature_cols = ["has_alert", "alert_count", "alert_type", "alert_streak", "alert_type_score", "alert_level_score"]

print(f"✅ Alert Features Added: {df_call.shape} (+{df_call.shape[1] - shape_before_alert_features[1]} cols)")
df_call[alert_feature_cols].sample(3)
```

# 신고 분류 기반 파생 변수

| 변수명 | 설명 |
|---|---|
| 구급 | 구급 신고 접수 여부 |
| 구조 | 구조 신고 접수 여부 |
| 기타 | 기타 신고 접수 여부 |
| 화재 | 화재 신고 접수 여부 |
| 부상 | 부상 신고 접수 여부 |
| 교통사고 | 교통사고 신고 접수 여부 |
| 업무운행 | 업무운행 신고 접수 여부 |
| 구급기타 | 구급기타 신고 접수 여부 |
| 자연재해 | 자연재해 신고 접수 여부 |

```python
shape_before_category_features = df_call.shape
```

```python
df_cat["tm"] = pd.to_datetime(df_cat["tm"])
df_cat["has_call"] = 1  # 단순히 존재했음을 의미

# 대분류(구급/구조/기타/화재) 기준 피벗
df_cat_cat = (
    df_cat.pivot_table(
        index=["tm", "address_gu", "sub_address"],
        columns="cat",
        values="has_call",
        aggfunc="max",
        fill_value=0
    )
    .reset_index()
)

# 소분류(부상/교통사고/업무운행/구급기타/자연재해) 기준 피벗
target_subcats = ["부상", "교통사고", "업무운행", "구급기타", "자연재해"]
df_cat_sub = (
    df_cat[df_cat["sub_cat"].isin(target_subcats)].pivot_table(
        index=["tm", "address_gu", "sub_address"],
        columns="sub_cat",
        values="has_call",
        aggfunc="max",
        fill_value=0
    )
    .reset_index()
)

df_cat_binary = df_cat_cat.merge(df_cat_sub, on=["tm", "address_gu", "sub_address"], how="outer").fillna(0)
```

```python
df_call = df_call.merge(df_cat_binary, on=["tm", "address_gu", "sub_address"], how="left")
```

```python
category_feature_cols = ["구급", "구조", "기타", "화재", "부상", "교통사고", "업무운행", "구급기타", "자연재해"]

print(f"✅ Category Features Added: {df_call.shape} (+{df_call.shape[1] - shape_before_category_features[1]} cols)")
df_call[category_feature_cols].sample(3)
```
